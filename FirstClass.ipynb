```python
# Rango de aÃ±os en pasos de cinco
year = [ii for ii in range(1950, 2006, 5)]
```

```python
C = []
value = 0.0
while type(value) == float:
    value = float(input("Insert value > "))
```

```python
C = []
for ii in range(len(year)):
    C.append(float(input("Insert value > ")))
```

```python
print(year, C)
```

```python
m = (C[-1]-C[0])/(year[-1]-year[0])
b = (year[-1]*C[0]-year[0]*C[-1])/(year[-1]-year[0])
print(m,b)
```

```python
import matplotlib.pyplot as plt 
```

```python
ymod = [m*x+b for x in year]
plt.plot(year, C, "o")
plt.plot(year, ymod)
```

```python
e = [(y-ym)/ym for y, ym in zip(C,ymod)]
```

```python
plt.plot(year, e, "o")
plt.axhline(0)
```
